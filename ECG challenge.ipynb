```python
import pyecg
from pyecg import ECGRecord
```

```python
record = ECGRecord.from_wfdb('ECG challenge\patient20\s0062lre.xyz')
```

```python
len(record)
```

```python
lead_list = ['i', 'ii', 'iii', 'avr', 'avl', 'avf', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'vx', 'vy', 'vz']
```

```python
lead_name = 'i'

time = np.array(record.time)
signal = np.array(record.get_lead(lead_name))
print(signal)
```

```python
signal.shape
```

```python
%matplotlib inline
plt.figure(figsize=(12,5))
plt.plot(time,signal)
```

```python
df = pd.DataFrame()
df = pd.concat([df,pd.Series(time),pd.Series(signal)], axis=1)
```

```python
for i in lead_list:
    path = "G:\(C) Program Files\MATLAB\\bin\patient20\s_0062_" + str(i) + ".csv"
    time = np.array(record.time)
    signal = np.array(record.get_lead(i))

    df = pd.DataFrame()
    df = pd.concat([df,pd.Series(time),pd.Series(signal)], axis=1)
    df.to_csv(path)

    print(path)
```
