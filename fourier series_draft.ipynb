```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_fourier_coefficient(X, Y, N_freq, T):
    A = []
    B = []
    for i in range(len(N_freq)):
        if i == 0:
            a = np.mean(Y)
            b = 0
        else:
            a = np.mean(Y*np.cos(N_freq[i]*np.pi*X/T)*2)
            b = np.mean(Y*np.sin(N_freq[i]*np.pi*X/T)*2)

        A.append(a)
        B.append(b)
    return A, B
```

```python
def fourier_series_spectrum(X, Y, N_freq, T):
    A, B = get_fourier_coefficient(X=X, Y=Y, N_freq=N_freq, T=T)
    F = []

    for i in range(len(N_freq)):
        f = np.sqrt(pow(A[i],2) + pow(B[i],2))
        F.append(f)
    return F
```

```python
def fourier_series(X, Y, N_freq, T):
    A, B = get_fourier_coefficient(X=X, Y=Y, N_freq=N_freq, T=T)
    F = []

    for x in X:
        f = 0
        for i in range(len(N_freq)):
            f = f + A[i]*np.cos(i*np.pi*x/T) + B[i]*np.sin(i*np.pi*x/T)
        F.append(f)
    return F
```

```python
# array initialize

Fs = 3000          # sampling frequency
Ts = 1/Fs          # time interval between samples
T = 2              # period of harmonic

x_start = 0
x_end = 4
X = np.linspace(x_start, x_end, num=(x_end - x_start)*Fs, endpoint=False)
N_freq = np.array(range(500))       # number of frequency components

F_domain = []
for n in N_freq:
    f = n/(T)
    F_domain.append(f)
```

```python
Y = []
for i in range(len(X)):
    if i%3000 < 1500:
        y = 1
    else: y = 0
    # y = np.sin(2*np.pi*7*X[i])
    # y = pow(np.e,-X[i])
    
    Y.append(y)
print(Y)
plt.plot(X,Y)
```

```python
F = fourier_series_spectrum(X, Y, N_freq, T/2)

plt.plot(F_domain,F)
```

```python
Z = fourier_series(X, Y, N_freq, T/2)

plt.plot(X, Z)
```
